**Задание 1** Определить результат выполнения операций над множествами. Считать элементы множества строками. 


```python
u = set("abcdefghijklmnopqrstuvwxyz")
A = {'b', 'f', 'g', 'm', 'o'}
B = {'b', 'g', 'h', 'l', 'u'}
C = {'e', 'f', 'm'}
D = { 'e', 'g', 'l', 'p', 'q', 'u', 'v'}

X = (A.difference(B)).union(C.intersection(D))
print("Ответ\n")
print(f"x = {X}")

A_u = u.difference(A)
B_u = u.difference(B)
Y = (A_u.intersection(B_u)).difference(C.union(D))

print(f"y = {Y}")
```
